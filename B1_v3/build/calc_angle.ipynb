```python
def gen_dir():
    M_PI=np.pi
    rand_var_1=np.random.uniform(low=0.0, high=1.0, size=None)**0.5
    rand_var_2=np.random.uniform(low=0.0, high=1.0, size=None)
    
    
    
    theta =rand_var_1*M_PI/512
    
    
    phi =  rand_var_2* M_PI*2
    
    return(theta,phi)

a=[]

for i in range(100000):
    theta,phi=gen_dir()
    xDirection = np.sin(theta) * np.cos(phi);
    yDirection = np.cos(theta);
    zDirection = np.sin(theta) * np.sin(phi);
    
    a.append([xDirection,yDirection,zDirection])
    
a=np.array(a)

plt.figure(figsize=(10,10))
plt.scatter(a[:,0],a[:,2],alpha=0.1,s=1)
plt.grid
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the starting point and components of the vector
origin = np.array([0, 0, 0])  # Vector origin point
vector = np.array([0, 0, 1])  # Vector components

# Set up the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vector from origin to the endpoint
ax.quiver(*origin, *vector, color='b', arrow_length_ratio=0.1)

# Set axis labels and limits
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Display the plot
plt.show()
```

```python
teta=np.pi/4
phi=0

R_z=np.array([[np.cos(phi),-np.sin(phi),0],
              [np.sin(phi), np.cos(phi),0],
              [0,0,1]])


R_y=np.array([[np.cos(teta),0,np.sin(teta)],
              [0, 1,0],
              [-np.sin(teta),0,np.cos(teta)]])


R=np.dot(R_z,R_y)





vector = np.array([0, 1, 0])

vector_rot=np.dot(R,vector)


# Set up the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vector from origin to the endpoint
ax.quiver(*origin, *vector, color='b', arrow_length_ratio=0.1)

ax.quiver(*origin, *vector_rot, color='r', arrow_length_ratio=0.1)
# Set axis labels and limits
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Display the plot
plt.show()
```
